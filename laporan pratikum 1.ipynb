```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("botak.csv")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["umur"], bins=30, kde=True, color="blue")
plt.xlabel("Umur")
plt.ylabel("Frekuensi")
plt.title("Distribusi Umur Responden")
plt.show()
```

```python
gender_counts = df["jenis_kelamin"].dropna().value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%", colors=["skyblue", "lightcoral"], startangle=140)
plt.title("Proporsi Jenis Kelamin Responden")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["gaji"], y=df["botak_prob"], alpha=0.5, color="green")
plt.xlabel("Gaji")
plt.ylabel("Probabilitas Kebotakan")
plt.title("Hubungan antara Gaji dan Probabilitas Kebotakan")
plt.show()
```
